# Logistic Regression and Accuracy

# implement LogisticRegression without using library

```python
import math

# Used Sigmoid function -- it converts any real value into a value between 0 and 1 -- represents probability -- values can be negative or large
# Sigmoid helps in binary classification

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Train Logistic Regression function -- Trains the Logistic Regression model
# Finds optimal weight and bias

def train_logistic_regression(X, y, lr, epochs):

    # Epochs define how many times the model iterates over the entire training dataset to minimize error.

    weight = 0
    bias = 0
    n = len(X)

    # Repeats learning process multiple times to reduce error.
    # Increasing epochs helps the model learn better by reducing error.
    
    for _ in range(epochs):
        for i in range(n):

            # Computes weighted sum of input.
            z = weight * X[i] + bias

            # Apply sigmoid -- Converts linear value to probability.
            y_pred = sigmoid(z)

        
            # Gradient calculation
            # calculate error -- Difference between predicted and actual output.
            error = y_pred - y[i]

            # Update weight -- Moves weight in direction of lower error.
            weight -= lr * error * X[i]

            # Update bias -- Adjusts bias to improve accuracy.
            bias -= lr * error

    return weight, bias

# Prediction function -- Uses trained model to make predictions 
# Converts probabilities into class labels (0 or 1)
def predict(X, weight, bias):
    predictions = []
    for x in X:

        # Linear calculation
        z = weight * x + bias
        y_pred = sigmoid(z)
        predictions.append(1 if y_pred >= 0.5 else 0)
    return predictions
```

```python
X = [1, 2, 3, 4, 5]
y = [0, 0, 1, 1, 1]

lr = 0.1
# full reading = 1 epoch
epochs = 1000

weight, bias = train_logistic_regression(X, y, lr, epochs)

print("Weight:", weight)
print("Bias:", bias)

predictions = predict(X, weight, bias)
print("Predictions:", predictions)
```
